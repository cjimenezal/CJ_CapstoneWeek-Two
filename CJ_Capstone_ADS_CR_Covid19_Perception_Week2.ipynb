## Introduction: Business Problem <a name="introduction"></a>

Background 
Costa Rica is a country that has a fairly high percentage of the population with sufficient education, the disease called COVID-19 brought, above all the Costa Rican people, great uncertainty, the government of the republic through the Ministry of Health, initially took a series determined of measurements.

Problem 
To measure the perception of the Costa Rican regarding the possibility of the risk of contagion with COVID-19 in all the country's provinces, the vast majority of the measures taken by the Government are known by Costa Ricans and are respected, however, there are some criticisms that they are not being fulfilled. This report aims to be one more tool, in this specific case to measure the perception of the population at the study sample, Costa Ricans aged between 18 to 80 years.

## Data: <a name="data"></a>

In this case there is no data to clean since they come from a survey carried out on the internet voluntarily, the study population was:
* Costa Ricans with ages between 18 to 80 years.
* Type of study: quantitative, designed with a structured questionnaire self-administered through the Internet through massive and open invitation, where the perception of contagion risk, knowledge and compliance with twenty-three measures established by the Government of Costa Rica and approximation were consulted of Costa Ricans facing COVID-19.
* Sample size: 797 Costa Ricans.
* Period: March 30 to 31, 2020.
* Type of sample: the sample is not entirely a random sample, for this reason, the database was weighted by gender, age, education, employment status, work sector and province of residence to restore the probability of selection in function of the structure of the Costa Rican population.
* Level of confidence in the estimates: 95%
* Profile of the sample: 3.5% (any difference in the figures greater than 3.5% shows a significant difference)

The original survey file carried out by UNED CR was summarized and its data summarized for a better understanding of the purpose of this report.
The data has been summarized and put in a CSV file in order for presentation purposes to be work with Pandas.

```python
# Loading Python Libraries.
import pandas as pd             # Importing Pandas primary data structure library
import numpy as np              # useful for many scientific computing in Python
import matplotlib.pyplot as plt # useful to make graphs
```

Now let's create a CR Covid-19 General Mesure Knowledge grid.

```python
# Load the General Measure csv file:
file1 =('//users/cjimenezal/Documents/Python Study/Visualization-withPython-Labs/Final-Assignment/MedidasXGeneral.csv')
# Read the csv file:
df_PCP=pd.read_csv(file1, index_col=0)

#using the method <code>head</code> to display the first 10 columns of the dataframe.
df_PCP.head(10)
```

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline

# Sorting the Dataframe
df_PCP.sort_values('Porcentaje Conocimiento Poblacion', ascending = True, inplace = True)

# Formatting to a percentage will be done in the plotting
df_PCP = df_PCP / 2233

#PLOTTING
ax1 = df_PCP.plot(kind = 'bar',
                figsize = (20,8),
                width = 0.8,
                color = ('#5cb85c', '#5bc0de', '#d9534f'),
                fontsize = 14)
ax1.set_title("Covid 19 Perception", 
             fontsize = 16)
ax1.legend(fontsize = 14)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)

# Adding the percentage values    
for p in ax1.patches:
    ax1.annotate("{:.02%}".format(p.get_height()),
                xy=(p.get_x()+0.02, p.get_height()+0.01))
```
